## Handling of Missing Values in Selected Input Features after Visualization

```python
def missing_values_handler(feature, dataframe):
    #
    dataframe[feature] = dataframe[feature].fillna('Unknown')
    
    ##Standardize the category of missing value with 'unknown'
    dataframe[feature] = dataframe[feature].replace(['na', 'unknown'], 'Unknown')
    
    #Replace with mode for input features that have realtively low missing values or is defined as ordinal data 
    if(feature == 'Type_of_collision' or feature == 'Area_accident_occured' or feature == 'Educational_level' or feature == 'Casualty_severity'):
        dataframe[feature] = dataframe[feature].replace('Unknown', dataframe[feature].mode()[0])
```

```python
for feature in features_after_visualization:
    missing_values_handler(feature, selected_features_visualization_df)
```

```python
#check out if all the missing values for each input features have handled
selected_features_visualization_df.info()
```

```python
def categorical_values_shortening(df, column, threshold):
    value_counts_series = df[column].value_counts()
    categorical_mapping = {}
    for i in range(len(value_counts_series)):
        if value_counts_series[i] >= threshold:
            categorical_mapping[value_counts_series.index[i]] = value_counts_series.index[i]
        else:
            categorical_mapping[value_counts_series.index[i]] = 'Other'
    df[column] = df[column].map(categorical_mapping)
```

```python
categorical_features_for_values_shortening = ['Type_of_collision',
                                'Area_accident_occured', 'Types_of_Junction', 
                                'Type_of_vehicle', 'Cause_of_accident', 
                               ]
for feature in categorical_features_for_values_shortening:
    categorical_values_shortening(selected_features_visualization_df, feature, 150)
```

## Ordinal Label Encoding for All Features

```python
from sklearn.preprocessing import OrdinalEncoder

educational_level_order = ['Illiterate', 'Writing & reading', 'Elementary school', 'Junior high school',
                           'High school', 'Above high school'
                          ]

oe_educational_level = OrdinalEncoder(categories=[educational_level_order])
selected_features_visualization_df['Educational_level'] = oe_educational_level.fit_transform(selected_features_visualization_df[['Educational_level']])

oe_nominal_categories = OrdinalEncoder()
label_encoded_nominal_features = ['Day_of_week', 'Type_of_collision',
                                'Area_accident_occured', 'Types_of_Junction',
                                 'Sex_of_driver', 'Age_band_of_driver',
                                'Driving_experience', 'Type_of_vehicle', 'Sex_of_casualty', 'Cause_of_accident', 
                                'Service_year_of_vehicle', 'Age_band_of_casualty', 'Vehicle_driver_relation'
                               ]
label_encoded_nominal_features_df = pd.DataFrame(oe_nominal_categories.fit_transform(selected_features_visualization_df[label_encoded_nominal_features]))
label_encoded_nominal_features_df.columns = selected_features_visualization_df[label_encoded_nominal_features].columns
label_encoded_nominal_features_df
```

```python
ordinal_and_numerical_features = ['Hour_of_Day', 'Casualty_severity', 'Number_of_vehicles_involved',
                                  'Number_of_casualties', 'Educational_level'
                                 ]
ordinal_and_numerical_features_df =  selected_features_visualization_df[ordinal_and_numerical_features]
encoded_df = pd.concat([ordinal_and_numerical_features_df, label_encoded_nominal_features_df], axis=1)
encoded_df.info()
```

## Label Encoding the Target Variable and Standardized all Features as Float type

```python
#Standardized all the values within the dataframe to integer data type
encoded_df['Casualty_severity'] = encoded_df['Casualty_severity'].astype('int')

from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
encoded_y = lb.fit_transform(y)
new_df['Accident_severity'] = encoded_y
new_df['Accident_severity']
#0 for fatal injury, 1 for serious injury, 2 for slight injury
```

```python
encoded_df.info()
```

## Feature Selection with Univariate Feature Selection of Chi2

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#carry out feature selections based on univariate statistical test of chi2
feature_selection = SelectKBest(score_func=chi2, k=9)
selected_features = feature_selection.fit_transform(encoded_df, new_df['Accident_severity'])
selected_features_names = feature_selection.get_feature_names_out()
```

```python
selected_features_names
```

```python
selected_features_univariate_df = pd.DataFrame(selected_features, columns=selected_features_names)
selected_features_univariate_df
```

## Oversampling of imbalanced Dataset

```python
from imblearn.over_sampling import SMOTENC

categorical_features_index = np.array(range(3, 9))

#Carry out oversampling via SMOTENC method
sm = SMOTENC(categorical_features=categorical_features_index, random_state=42)
x_res, y_res = sm.fit_resample(selected_features_univariate_df, new_df['Accident_severity'])
```

```python
y_res.value_counts()
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, random_state=42)
rf = RandomForestClassifier(n_estimators=500)
rf.fit(x_train, y_train)
ypred = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, ypred))
```
